## Using {Matplotlib.pyplot + numpy}

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Plotting x, y

```python
x = np.array([3, 6])
y = np.array([0, 10])

plt.plot(x, y)

plt.show()
```

### Plotting multiple data points

```python
x = np.array([0, 1, 2, 3, 4])
y = np.array([5, 10, 0, 50, 20])

plt.plot(x, y)

plt.show()
```

### Plotting multiple lines

```python
y1 = np.array([1, 3, 1, 7])
y2 = np.array([3, 1, 7, 1])

plt.plot(y1)
plt.plot(y2)

plt.show()
```

### Default X-Points

```python
y = np.array([10, 6, 7, 1, 9])

plt.plot(y)

plt.show()
```

### Plotting x, y dots

```python
x = np.array([3, 6])
y = np.array([0, 10])

plt.plot(x, y, 'o')

plt.show()
```

## Using Markers

<br>'o' = Circle
<br>'*' = Star
<br>'.' = point
<br>
<br>'+' = Plus
<br>'P' = {filled} Plus
<br>'x' = X
<br>'X' = {filled} X
<br>'d' = Diamond 
<br>'D' = {bold} Diamond
<br>'h' = Hexagon
<br>'H' = {bold} Hexagon
<br>
<br>'s' = Square
<br>'p' = Pentagon
<br>
<br>'v' = Triangle Down
<br>'^' = Triangle Up
<br>'<' = Triangle Left
<br>'>' = Triangle Right

```python
y = np.array([1, 11, 3, 13, 5, 15])

plt.plot(y, marker = 'D')

plt.show()
```

### Maker Size + color

```python
y = np.array([1, 11, 3, 13, 5, 15])

plt.plot(y, marker='D', ms = 15, mec = 'm', mfc = 'w')

plt.show()
```

## Line Style + Size + Color 

```python
y = np.array([1, 11, 3, 13, 5, 15])

plt.plot(y, linestyle = "dotted", color = "black", linewidth = '5')

plt.show()
```

## Using FMT

<h5>Line Reference</h5>
'-'  = Solid line
<br>':'  = Dotted line
<br>'--' = Dashed line
<br>'-.' = Dashed/dotted line

<h5>Color Reference</h5>
'r' = Red
<br>'g' = Green
<br>'b' = Blue
<br>'c' = Cyan
<br>'m' = Magenta
<br>'y' = Yellow
<br>'k' = Black
<br>'w' = White

```python
y = np.array([1, 11, 3, 13, 5, 15])

plt.plot(y, 'D-.b')

plt.show()
```
